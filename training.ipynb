```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv("datasets/deliveries.csv")
df
```

```python
df
```

```python
df.loc[df['wicket_type'].notnull() & df['fielder'].isnull(), 'fielder'] = "no fielder"
df[df['fielder']=="no fielder"].head()
```

```python
df2= df.copy()
df2.isnull().sum()
```

```python
df2.loc[df2['wicket_type'].isnull(), 'wicket_type'] = "not out"
df2.loc[df2['player_dismissed'].isnull(), 'player_dismissed']= "not out"
df2.loc[df2['fielder'].isnull(), 'fielder'] = "not out"
df2.isnull().sum()
```

```python
maindf= df2.copy()
maindf.drop(columns=['match_no', 'date'], inplace=True)
```

```python
for i in range(len(maindf)):
    if int(maindf.loc[i, 'over'])<6:
        maindf.loc[i, 'over'] = "powerplay"
    elif int(maindf.loc[i, 'over'])<16:
        maindf.loc[i, 'over'] = "middle overs"
    else:
        maindf.loc[i, 'over'] = "death overs"

    
```

```python
maindf
```

```python
pruns=0
mruns=0
druns=0
for i in range(len(maindf)):
    if maindf.loc[i, 'over'] == "powerplay":
        pruns += maindf.loc[i, 'runs_of_bat']
    elif maindf.loc[i, 'over'] == "middle overs":
        mruns += maindf.loc[i, 'runs_of_bat']
    else:
        druns += maindf.loc[i, 'runs_of_bat']
print("Powerplay Runs:", pruns)
print("Middle Overs Runs:", mruns)
print("Death Overs Runs:", druns)
```

```python
#plotting a pie chart
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.pie([pruns, mruns, druns], labels=['Powerplay', 'Middle Overs', 'Death Overs'], autopct='%1.1f%%', startangle=140)
plt.title('Runs Scored in Different Phases of the Game')
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is circular.
plt.show()
#saviing the image
plt.savefig("fancy stats/runs_scored_phases.png")
```

```python
team_wise_runs={}
for i in range(len(maindf)):
    team = maindf.loc[i, 'batting_team']
    if team not in team_wise_runs:
        team_wise_runs[team] = 0
    team_wise_runs[team] += int(maindf.loc[i, 'runs_of_bat'])

team_wise_runs
```

```python
#plotting a bar chart in descending order of the runs scored by each team
team_wise_runs = dict(sorted(team_wise_runs.items(), key=lambda item: item[1], reverse=True))
plt.figure(figsize=(12, 6))
plt.bar(team_wise_runs.keys(), team_wise_runs.values(), color='blue', alpha=0.7)
plt.xlabel('Teams')
plt.ylabel('Runs Scored')
plt.title('Total Runs Scored by Each Team')
plt.xticks(rotation=45)
plt.tight_layout()  # Adjust layout to prevent clipping of tick-labels
plt.show()
#saving the image
plt.savefig("fancy stats/team_wise_runs.png")
```

```python
sixes_by_each_team={}
for i in range(len(maindf)):
    team = maindf.loc[i, 'batting_team']
    if maindf.loc[i, 'runs_of_bat'] == 6:
        if team not in sixes_by_each_team:
            sixes_by_each_team[team] = 0
        sixes_by_each_team[team] += 1
sixes_by_each_team = dict(sorted(sixes_by_each_team.items(), key=lambda item: item[1], reverse=True))
sixes_by_each_team
```

```python
#plotting bar graph for no of sixes
plt.figure(figsize=(12, 6))
plt.bar(sixes_by_each_team.keys(), sixes_by_each_team.values(), color='green', alpha=0.7)
plt.xlabel('Teams')
plt.ylabel('Number of Sixes')
plt.title('Number of Sixes by Each Team')
plt.xticks(rotation=45)
plt.tight_layout()  # Adjust layout to prevent clipping of tick-labels
plt.show()
#saving the image
plt.savefig("fancy stats/sixes_by_each_team.png")
```

```python
#tuk tuk batters
tuk_tuk_batters = {}
for i in range(len(maindf)):
    batter = maindf.loc[i, 'striker']
    if batter not in tuk_tuk_batters:
        tuk_tuk_batters[batter] = [0,0]
    tuk_tuk_batters[batter][0] += int(maindf.loc[i, 'runs_of_bat'])
    tuk_tuk_batters[batter][1] += 1
tuk_tuk_strike_rates={}
for i in tuk_tuk_batters:
    if tuk_tuk_batters[i][1] <10:
        continue
    tuk_tuk_strike_rates[i] = (tuk_tuk_batters[i][0] / tuk_tuk_batters[i][1]) * 100
tuk_tuk_strike_rates = dict(sorted(tuk_tuk_strike_rates.items(), key=lambda item: item[1]))
tuk_tuk_strike_rates
        
```
